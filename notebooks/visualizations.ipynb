```python
from matplotlib import pyplot as plt 
import seaborn as sb 
import pandas as pd
import numpy as np
```

```python
# set visualization styles
sb.set(style="white", font_scale=1.2)
# plt.style.use('ggplot')
sb.set(font='serif')
sb.set_palette("deep")
sb.set_context("paper")

plt.rcParams['xtick.bottom'] = True
plt.rcParams['ytick.left'] = True

plt.rcParams['font.family'] = 'serif'
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10

plt.rcParams['axes.facecolor']='w'
```

```python
base_results = {
    "Raw prompt": 16.79,
    "Zero-shot CoT": 50.78,
    "CoT": 55.45,
    "Plan & Solve": 49.80,
}
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
ax.bar(base_results.keys(), base_results.values(), color=sb.xkcd_rgb["flat blue"])
ax.set_ylabel("Solve rate (%)")
ax.set_ylim(0, 60)

ax.spines['bottom'].set_color('0.5')
ax.spines['left'].set_color('0.5')

# mark bars with values rounded
for i, v in enumerate(base_results.values()):
    ax.text(i, v + 1, str(round(v)) + "%", ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.savefig("base_results.pdf")
plt.show()
```
